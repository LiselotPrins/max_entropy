From the theory we know there does not exist a MaxEnt distribution for: 
support $S=\mathbb{R}$ and contraints on the first third moments (or, 
equivalently, the mean, variance and skewness).

So I was curious was my algorithm would do with this case.
Surprisingly, it's very easy to have the algorithm converge.
This is probably due to the approximation of the domain as a finite interval.
In cases of non-zero third moment, we see a 'bump' at the edges of the support.
This goes further out as the size of the support is increased. This may 
have something to do with there existing no MaxEnt distribution on the whole of 
$\mathbb{R}$.

```python
import numpy as np
from autograd import numpy as anp
import  matplotlib.pyplot as plt
from python_files.maxent import MaxEnt
import python_files.densities as d
import python_files.figure_settings
```

```python
n = 350
three_moments = [0,1,9]
```

```python
l,u = -6,6
ob_skewkurt = MaxEnt(support=(l,u),
                    n_quadrature=n,
                    b_constraints=d.b_constraint_three_moments(*three_moments),
                    f_vector=d.f_constraint_three_moments,
                    k_max=400)

# ob_skewkurt.visualize_algorithm()
ob_skewkurt.visualize_solution(
    title=f"MaxEnt for $(\\mu_1,\\mu_2,\\mu_3)=({three_moments[0]},{three_moments[1]},{three_moments[2]})$ & $(l,u)=({l},{u})$")
```
